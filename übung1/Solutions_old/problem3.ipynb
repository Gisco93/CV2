We start by importing packages for manipulating and displaying images and plotting.

```julia
using Images, ImageView
import PyPlot
```

Let's load the image ...

```julia
img = load("../data/a1p3.png")
```

... and convert it to grayscale.

```julia
imgg = Gray.(img)
```

We need to convert the image to floating point format:

```julia
imgf = float64.(imgg);
f = channelview(imgf);
```

Display the image using ImageView.

```julia
imshow(imgf, name="ImageView")
```

Notice how it spawns a new window?
Now let's do the same with PyPlot, this time with a nice plot within the notebook.
(It will spawn a window though if you are running the command outside of IJulia)

```julia
PyPlot.imshow(f, "gray")
PyPlot.axis("off")
PyPlot.title("PyPlot")
```

Finally, let's compute some low-level image statistics.

```julia
m0 = minimum(f)
m1 = sum(f)./length(f)
m2 = maximum(f)
println(@sprintf "min: %.2f, mean: %.2f, max:%.2f" m0 m1 m2)
```
